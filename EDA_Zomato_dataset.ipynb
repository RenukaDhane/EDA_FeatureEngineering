```python
df_country=pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
country_values
```

```python
# top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
plt.show()
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head(3)
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

## Observations
1. when rating between 4.5 to 4.9 --> Excellent
2. when rating between 4.4 to 4.0 --> Very Good
3. when rating between 3.9 to 3.5 --> Good
4. when rating between 3.0 to 2.5 --> Average
5. when rating between 2.4 to 1.8 --> Poor


```python

ratings['Rating text'].value_counts().index
```

```python

plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
plt.show()
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

Observations
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.xlabel('Rating Color')
plt.title('Color of ratings')
plt.show()
```
